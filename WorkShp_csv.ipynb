```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("/content/Startup.csv")

data["year"] = pd.to_numeric(data["year"], errors="coerce")
data["amount"] = pd.to_numeric(data["amount"], errors="coerce")

data = data.dropna(subset=["year", "amount"])

investment_per_year = data.groupby("year")["amount"].sum().reset_index()

plt.figure(figsize=(10, 6))

plt.plot(
    investment_per_year["year"],
    investment_per_year["amount"],
    marker="o",
    markersize=8,
    linestyle="--",
    linewidth=3,
    label="Total Amount"
)

plt.title("Year Wise Amount Investment", color="green", fontsize=25)

plt.xlabel("Year", fontsize=14)
plt.ylabel("Investment Amount", fontsize=14)

plt.grid(True, linestyle="--")
plt.legend()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("/content/Startup.csv")

data["city"] = data["city"].astype(str).str.strip()

city_startups = data["city"].value_counts().head(10)

colors = [
    "pink", "black", "green", "orange", "purple",
    "cyan", "magenta", "yellow", "brown", "pink"
]

plt.figure(figsize=(10, 6))

plt.bar(
    city_startups.index,
    city_startups.values,
    color=colors,
    edgecolor="black",
    linewidth=1.5
)

plt.title("Top 10 Cities Highest Number of Startups", color="black", fontsize=20)
plt.xlabel("Cities", fontsize=14)
plt.ylabel("Number of Startups", fontsize=14)

plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--")

plt.show()
```

```python
Demo = data["vertical"].value_counts().head(10).sort_values()

colors = [
    "black", "yellow", "green", "brown", "pink",
    "purple", "red", "blue", "orange", "cyan"
]

plt.figure(figsize=(10, 6))

plt.barh(
    Demo.index,
    Demo.values,
    color=colors,
    edgecolor="black"
)

plt.title("Top 10 Startup Verticals", fontsize=20)
plt.xlabel("Number of Startups", fontsize=14)
plt.ylabel("Vertical", fontsize=14)

plt.grid(axis="x", linestyle="--")

plt.show()
```

```python
import matplotlib.pyplot as plt

Demo = data["round"].value_counts().head(10)

colors = [
    "red", "blue", "green", "orange", "purple",
    "cyan", "magenta", "yellow", "brown", "pink"
]

plt.figure(figsize=(14, 8))

bars = plt.bar(
    Demo.index,
    Demo.values,
    color=colors,
    edgecolor="black",
    linewidth=2
)

plt.title("Top 10 Funding Rounds", fontsize=22, color="black")
plt.xlabel("Funding Round", fontsize=16)
plt.ylabel("Number of Startups", fontsize=16)

plt.xticks(rotation=30, fontsize=12)
plt.yticks(fontsize=12)

plt.grid(axis="y", linestyle="--")

for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height(),
        str(int(bar.get_height())),
        ha="center",
        va="bottom",
        fontsize=11
    )

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

Demo = data["round"].value_counts().head(7)

colors = [
    "#4E79A7",
    "#F28E2B",
    "#E15759",
    "#76B7B2",
    "#59A14F",
    "#EDC948",
    "#B07AA1"
]

plt.figure(figsize=(10, 8))

plt.pie(
    Demo.values,
    labels=Demo.index,
    colors=colors,
    autopct="%1.1f%%",
    startangle=140,
    explode=[0.08, 0.05, 0.05, 0.03, 0.03, 0.03, 0.03],
    shadow=True,
    textprops={"fontsize": 12, "fontweight": "bold"},
    wedgeprops={"edgecolor": "black", "linewidth": 1.5}
)

plt.title("Top 7 Funding Rounds Distribution", fontsize=22, fontweight="bold")
plt.axis("equal")
plt.tight_layout()

plt.show()
```

```python
# How are startup founding amounts distributed?

import matplotlib.pyplot as plt

Demo = data["amount"].dropna()

plt.figure(figsize=(10, 7))

plt.hist(
    Demo,
    bins=10,
    color="skyblue",
    edgecolor="black",
    linewidth=1.5
)

plt.title("Distribution of Startup Funding Amounts", fontsize=22, fontweight="bold")
plt.xlabel("Funding Amount", fontsize=15)
plt.ylabel("Number of Startups", fontsize=15)

plt.grid(axis="y", linestyle="--")

plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt

data["year"] = data["year"].astype(int)
data["amount"] = pd.to_numeric(data["amount"], errors="coerce")

plt.figure(figsize=(10, 6))

plt.scatter(
    data["year"],
    data["amount"],
    color="blue",
    edgecolors="black",
    alpha=0.7
)

plt.title(" Funding Year and Funding Amount", fontsize=20)
plt.xlabel("Funding Year", fontsize=14)
plt.ylabel("Funding Amount", fontsize=14)

plt.grid(linestyle="--")

plt.show()
```
